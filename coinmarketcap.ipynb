# Next: (1) Scrape historic price 1M,3M,6M,1Y,2Y

```python
records = []
n=0
table0=table[0]
for tr in table0.findAll("tr"):
    trs = tr.findAll("td")
    if len(trs)>0:
        record = []
        record.append(trs[1].a["href"])
        record.append(trs[2].text)
        records.append(record)
        n=n+1
```

```python
coins = pd.DataFrame(data=records, columns=['Link','Tkr'])
coins
```

```python
import datetime
# from datetime import datetime
```

```python
# for i in coins.iterrows():
clink=coins['Link'][2]
url10="https://coinmarketcap.com{}historical-data/?start=20130428&end=20180406".format(clink)
url1 = urllib2.urlopen(url10)
content1 = url1.read()
soup1 = BeautifulSoup(content1)
table1=str(soup1.find_all("table")[0])
df3=pd.read_html(table1)[0]
df33=pd.DataFrame(df3)
df33['Date']=pd.to_datetime(df33['Date'], infer_datetime_format=True)
```

```python
date_0=df33.iloc[-1]['Date']
date_1M = date_0 + datetime.timedelta(days=30)
date_3M = date_0 + datetime.timedelta(days=90)
date_6M = date_0 + datetime.timedelta(days=180)
date_1Y = date_0 + datetime.timedelta(days=360)
date_2Y = date_0 + datetime.timedelta(days=720)
```

```python
#Init values
a=df33.loc[df33['Date'] == date_0]
a=a.drop('Date',1)
a=a.reset_index(drop=True)
#1M values
b=df33.loc[df33['Date'] == date_1M]
b=b.drop('Date',1)
b=b.reset_index(drop=True)
#3M values
c=df33.loc[df33['Date'] == date_3M]
c=c.drop('Date',1)
c=c.reset_index(drop=True)
#6M values
d=df33.loc[df33['Date'] == date_6M]
d=d.drop('Date',1)
d=d.reset_index(drop=True)
#1Y values
e=df33.loc[df33['Date'] == date_1Y]
e=e.drop('Date',1)
e=e.reset_index(drop=True)
#2Y values
f=df33.loc[df33['Date'] == date_2Y]
f=f.drop('Date',1)
f=f.reset_index(drop=True)
data33 = pd.concat([a,b,c,d,e,f], axis=1)
```
